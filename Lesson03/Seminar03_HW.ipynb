## Задание 3. Составьте несколько сводных таблиц.

Для дальнейшего анализа и получения корректных результатов необходимо очистить данные от дублей. 

```python
df['rank'] = df.groupby('id')['date'].rank(ascending=False)
df['rank'].value_counts()
```

```python
df[df['id'].isin(set(df[df['rank'] > 1]['id']))].sort_values(['id','date'])
```

Дублирующиеся строки по домам отличаются датой и ценой продажи. Для анализа возьмем последние данные (с максимальной датой продажи).

```python
df = df[df['rank'] == 1].iloc[:, :-1]
df
```

### 3.1. Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей.

```python
df.pivot_table(
    index = 'bedrooms',
    values = 'price',
    aggfunc = 'mean'
).round({'price':1}).sort_values('price')
```

### 3.2. Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.pivot_table(
    index = 'condition',
    aggfunc = {'price': ['min', 'mean', 'max']}
).round({('price',  'min'):1, ('price', 'mean'):1, ('price',  'max'):1})
```

### 3.3. Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида.

```python
df.pivot_table(
    index='waterfront',
    columns='view',
    values='id',
    aggfunc='count',
    fill_value = '--'
)
```

### 3.4. Каких домов в зависимости от этажности и количества спален больше?

```python
def red_max(x):
    return np.where(x == np.max(x.to_numpy()), f"color: red ;", None)


df.pivot_table(
    index='floors',
    columns='bedrooms',
    values='id',
    aggfunc='count',
    fill_value = 0
).style.apply(red_max, axis=None)
```

### 3.5. Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома.

```python
df.pivot_table(
    index='grade',
    columns='condition',
    values='price',
    aggfunc='median',
    fill_value = '-'
)
```
